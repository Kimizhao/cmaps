## 制作rgb文件步骤
1. 转换工具转换成rgb文件：
convert目录下
- json2rgb.py json文本转换为rgb
- txt2rgb.py text文件转换为rgb

2. 制作后的*.rgb拷贝到cmaps/colormaps/self_defined

3. 重新安装cnmaps
```shell
$ pip uninstall cmaps
$ python setup.py install
```
